2. Note that there should only be three separate countries. Eliminate all variations, so that 'South Africa', 'United Kingdom' and 'United States' are the only three countries.

```python

# Example of DataFrame for demonstration
data = {
    'country': ['South Africa','south africa','United Kindom','UK','united states','USA','U.S.'],
    'values' : [1, 2, 3, 4, 5, 6, 7]
}
df = pd.DataFrame(data)

# List of standardized country names
standard_countries = ['South Africa','United Kingdom','United States']
from fuzzywuzzy import process

def get_closest_country(country, choices):
    closest_match = process.extractOne(country, choices)
    return closest_match[0] if closest_match[1] > 80 else country

# Apply the function to the 'country' column
df['country'] = df['country'].apply(lambda x:get_closest_country(x, standard_countries))

# Filter the DataFrame to include only the three countries
df = df[df['country'].isin(standard_countries)]

# verify the change
print(df['country'].unique())

```

3. Create a new column called `days_ago` in the DataFrame that is a copy of the 'date_measured' column but instead it is a number that shows how many days ago it was measured from the current date. Note that the current date can be obtained using `datetime.date.today()`.

```python
import pandas as pd
from datetime import datetime
from datetime import date

# DataFrame for demonstration
data = {
    'country': ['South Africa','south africa','United Kindom','UK','united states','USA','U.S.'],
    'values' : [1, 2, 3, 4, 5, 6, 7],

    'date_measured':['2023-06-01', '2023-06-02', '2023-06-03', '2023-06-04', '2023-06-05', '2023-06-06', '2023-06-07']
}
df = pd.DataFrame(data)

# Convert date_measured to datetime 
df['date_measured'] = pd.to_datetime(df['date_measured'])

# Get the current date
current_date = date.today()

# Calculate the difference in days and create the new column
df['days_ago'] = (current_date - df['date_measured'].dt.date).apply(lambda x: x.days)

# Verify new the column
print(df)
```
